```python
import os
import argparse

# third party
import numpy as np
import nibabel as nib
import torch
import json
import time
import neurite as ne
import pystrum

# import voxelmorph with pytorch backend
os.environ['NEURITE_BACKEND'] = 'pytorch'
os.environ['VXM_BACKEND'] = 'pytorch'
import voxelmorph as vxm   # nopep8

device = 'cpu'
os.environ['CUDA_VISIBLE_DEVICES'] = '-1'
add_feat_axis = True
eval_set = "training_paired_images"
json_file = '/home/shreya/scratch/voxelmorph/images/AbdomenMRCT/AbdomenMRCT_dataset.json'
with open(json_file) as f:
    pairs = json.load(f)[eval_set]

pair = pairs[0]
```

```python
moving = vxm.py.utils.load_volfile(pair["1"], add_batch_axis=True, add_feat_axis=add_feat_axis)
fixed, fixed_affine = vxm.py.utils.load_volfile(pair["0"], add_batch_axis=True, add_feat_axis=add_feat_axis, ret_affine=True)
moving_seg = vxm.py.utils.load_volfile(pair["1"].replace('imagesTr','labelsTr'), add_batch_axis=True, add_feat_axis=add_feat_axis,normalize=False)
fixed_seg = vxm.py.utils.load_volfile(pair["0"].replace('imagesTr','labelsTr'), add_batch_axis=True, add_feat_axis=add_feat_axis,normalize=False)
```

```python
args_model = '/home/shreya/scratch/voxelmorph/models/20240131023912/0180.pt'
```

```python
inshape = moving.shape[1:-1]
print(inshape)

# load and set up model
model = vxm.networks.VxmDense.load(args_model, device)
model.to(device)
model.eval()
```

```python
# set up tensors and permute
input_moving = torch.from_numpy(moving).to(device).float().permute(0, 4, 1, 2, 3)
input_fixed = torch.from_numpy(fixed).to(device).float().permute(0, 4, 1, 2, 3)
moving_seg = torch.from_numpy(moving_seg).to(device).float().permute(0, 4, 1, 2, 3)
fixed_seg = torch.from_numpy(fixed_seg).to(device).float().permute(0, 4, 1, 2, 3)
```

```python
moved, warp = model(input_moving, input_fixed, registration=True)
```

```python
transformer = vxm.layers.SpatialTransformer(inshape)
warped_seg = transformer(moving_seg, warp)
```

```python
mid_slices_fixed = [np.take(input_fixed.detach().squeeze().squeeze(), inshape[d]//2, axis=d) for d in range(3)]
mid_slices_fixed[1] = np.rot90(mid_slices_fixed[1], 1)
mid_slices_fixed[2] = np.rot90(mid_slices_fixed[2], -1)

mid_slices_pred = [np.take(moved.detach().squeeze().squeeze(), inshape[d]//2, axis=d) for d in range(3)]
mid_slices_pred[1] = np.rot90(mid_slices_pred[1], 1)
mid_slices_pred[2] = np.rot90(mid_slices_pred[2], -1)
ne.plot.slices(mid_slices_fixed + mid_slices_pred, cmaps=['gray'], do_colorbars=True, grid=[2,3]);
```
